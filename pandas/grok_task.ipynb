```python
#Create a new column categorizing countries into ”Low”, ”Medium”, and ”High”
#GDP based on quartiles of the gdp column.
df["gdp_category"] = pd.qcut(df["gdp"], q=3, labels=["Low", "Medium", "High"])
df[["country","gdp","gdp_category"]].head(25)
```

```python
#For each continent, calculate the ratio of co2_emissions to population
#and identify the country with the highest ratio.
continent_ratio = df.groupby("continent")[["co2_emissions", "population"]].sum()
continent_ratio["co2_per_capita"] = continent_ratio["co2_emissions"] / continent_ratio["population"]
continent_ratio

df["co2_per_capita"] = df["co2_emissions"] / df["population"]
highest_per_country = df.loc[df.groupby("continent")["co2_per_capita"].idxmax()]
highest_per_country[["continent","country"]]
```

```python
#Use a rolling window of 5 countries (sorted by population) to compute the
#average life_expectancy.
df.sort_values("population",inplace=True)
df["life_expectancy"].rolling(5,min_periods=1).mean()
```

```python
#Apply a custom function to normalize health_expenditure_capita within
#each region to a 0–1 scale.
df["normalized"] = df.groupby("region")["health_expenditure_capita"].transform(lambda x: (x - x.min()) / (x.max() - x.min()))
df[["country","normalized"]]
```

```python
#find countries where the urban_population is more than 50% of the total
#population and electricity_access_pct is below 90.
df.query("urban_population > population / 2 and electricity_access_pct < 90")
```

```python
# Group by democracy_type and compute the average gdp only for countries
# with non-missing inflation values.
df_filtered = df[df["inflation"].notna()]  # Keep rows where inflation is NOT missing
df_filtered.groupby("democracy_type")["gdp"].mean()
```

```python
# Create a new DataFrame showing the top 5 countries per continent by internet_pct,
# handling ties appropriately.
df['rank'] = df.groupby('continent')['internet_pct'].rank(method='min', ascending=False)
top5_per_continent = df[df['rank'] <= 5].copy()
top5_per_continent
```

```python
#Identify outliers in co2_emissions using the IQR method and list the countries with these outliers
Q1 = df['co2_emissions'].quantile(0.25)
Q3 = df['co2_emissions'].quantile(0.75)
print(Q3)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df['co2_emissions'] < lower_bound) | (df['co2_emissions'] > upper_bound)]
outlier_countries = outliers[['country', 'co2_emissions']]
outlier_countries
```

```python
#For each region, calculate the percentage contribution of each country’s gdp
#to the region’s total gdp.
df['gdp_pct'] = df['gdp'] / (df.groupby('region')['gdp'].transform('sum')) * 100
for region, group in df.groupby('region'):
    print(f"\n--- {region} ---")
    print(group[['country', 'region',"gdp",'gdp_pct']])
```

```python
# Perform a time-series-like analysis by assuming birth_rate and death_rate
# are annual metrics and estimate population growth rate.
df["growth_rate"] = ((df["birth_rate"] - df["death_rate"]) / 1000) * 100

result = df[["country", "birth_rate", "death_rate", "growth_rate"]]
print(result)
```

```python
# Apply a custom function to rank countries within each continent by gdp and
# add the rank as a new column
import pandas as pd

def rank_gdp(series):
    return series.rank(ascending=False, method='min')

df['gdp_rank'] = df.groupby('continent')['gdp'].transform(rank_gdp)

print(df[['country', 'continent', 'gdp', 'gdp_rank']])
```
