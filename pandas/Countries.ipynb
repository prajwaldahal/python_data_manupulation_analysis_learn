```python
#total region
df["region"].value_counts().count()
```

```python
#how many coutries lies in eastern europe
df["region"].value_counts()["Eastern Europe"]
```

```python
df[df["region"]=="Eastern Europe"]["country"]
```

```python
#political leader of 2nd largest popn country
df[df["population"]==df["population"].nlargest(3).iloc[1]]["political_leader"]
```

```python
#coutries with no political_leader
df[df["political_leader"].isna()].count()
```

```python
#republic countries
def republicCount(txt):
    if "republic" in txt.lower():
        return True
    return False
```

```python
df[df["country_long"].apply(republicCount)]["country"].count()
```
