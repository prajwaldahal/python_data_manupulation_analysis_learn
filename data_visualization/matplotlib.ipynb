```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Matplot basic

```python
x=np.linspace(0,5,11)
x
y=x**3
y
```

```python
plt.plot(x,y)
plt.title("line graph demo")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```

SUBPLOTS

```python
plt.subplot(2,2,1)
plt.plot(x**2,y/2)
plt.subplot(2,2,2)
plt.plot(y/2,x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
fig=plt.figure(figsize=(12,8),dpi=100)

axis=fig.add_axes([0.1,0.1,1,1])
axis.plot(x,y)
axis2=fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
```
